```python
#separate features and target
data = iris.values
X = data[:, 0:4]
Y= data[:, 4]
print(X)
```

```python
print(Y)
```

```python
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()
# transform the string labels to integer
iris['species'] = le.fit_transform(iris['species'])
iris.head()
```

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(X,Y,test_size=0.3, shuffle = True, random_state = 69)
```

```python

x_train.shape
```

```python
x_test.shape
```

```python
y_train.shape
```

```python
y_test.shape
```

```python
from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
x_train = sc.fit_transform(x_train)
x_test = sc.transform(x_test)
print(x_train[:1])
```

```python
from sklearn.svm import SVC

model_svc= SVC()
model_svc.fit(x_train, y_train)
```

```python
y_pred= model_svc.predict(x_test)
y_pred
```

```python
# to find the accuracy of the model and view the confusion matrix
from sklearn.metrics import accuracy_score,confusion_matrix
confusion_matrix(y_test, y_pred)
```

```python
#accuracy
accuracy=accuracy_score(y_test,y_pred)*100
print("Accuracy of the model is {:.2f}".format(accuracy))
```

```python
from sklearn.linear_model import LogisticRegression

model_lgr = LogisticRegression()
model_lgr.fit(x_train,y_train)
```

```python
y_pred = model_lgr.predict(x_test)
y_pred
```

```python
# to view the confusion matrix of the model
confusion_matrix(y_test, y_pred)
```

```python
#accuracy
accuracy=accuracy_score(y_test,y_pred)*100
print("Accuracy of the model is {:.2f}".format(accuracy))
```

```python
from sklearn.tree import DecisionTreeClassifier
model_DTC = DecisionTreeClassifier()
model_DTC.fit(x_train, y_train)
```

```python
y_pred = model_DTC.predict(x_test)
confusion_matrix(y_test, y_pred) #confusion matrix of the model
```

```python
#accuracy
accuracy=accuracy_score(y_test,y_pred)*100
print("Accuracy of the model is {:.2f}".format(accuracy))
```

```python
from sklearn.neighbors import KNeighborsClassifier
model_knn = KNeighborsClassifier()
model_knn.fit(x_train, y_train)
```

```python
y_pred = model_knn.predict(x_test)
y_pred
```

```python
from sklearn.linear_model import LogisticRegression
model = LogisticRegression()
model.fit(x_train, y_train)
y_pred = model.predict(x_test)
accuracy = accuracy_score(y_test, y_pred)
print("Accuracy: ", accuracy * 100)
```
